```python
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from datasets import load_dataset
from datasets import Dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
```

```python
class load_detector:
    def __init__(self):
        dataset = load_dataset("ealvaradob/phishing-dataset", "texts", trust_remote_code=True)
        df = dataset['train'].to_pandas()
        x = df['text']
        y = df['label']

        train, test = train_test_split(df, test_size=0.2, random_state=42)
        train, test = Dataset.from_pandas(train, preserve_index=False), Dataset.from_pandas(test, preserve_index=False)
        
        # Convert text data to TF-IDF features
        vectorizer = TfidfVectorizer()

        # Create a pipeline that first transforms the data and then fits the classifier
        self.pipeline = make_pipeline(vectorizer, RandomForestClassifier())

        # Fit the pipeline
        self.pipeline.fit(x, y)
    
    def is_phish(self, phish):
        self.phish = phish
        # text you want to make prediction on
        new_data = [self.phish]

        # Transform the new data using the vectorizer
        new_data_transformed = self.pipeline.named_steps['tfidfvectorizer'].transform(new_data)

        # Make predictions
        predictions = self.pipeline.named_steps['randomforestclassifier'].predict(new_data_transformed[0])

        if predictions[0] == 0:
            print("Your message is safe")
        else:
            print("Your message is unsafe")
        
```

```python
load = load_detector()
```

```python
predict = load.is_phish("Can we go out tonight?")
```
